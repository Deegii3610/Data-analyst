```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## College


```python

college = pd.read_csv('College.csv')

college.head()
```

```python
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
```

```python
college2.head()
```

```python
college3.head()
```

```python
college = college3
college.head(2)
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]], figsize=(10,10))
```

```python
sns.boxplot(x='Private', y='Outstate', data=college)
plt.title('Outstate vs Private')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])

college['Elite'].head()
```

```python
sns.boxplot(x='Elite', y='Outstate', data=college)
plt.title('Elite vs Private')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))# sharex=True
axs[0, 0].hist(college["Outstate"], bins=20, color='blue', alpha=0.7)
axs[0, 0].set_title('Outstate')

axs[0, 1].hist(college["Apps"], bins=20, color='green', alpha=0.7)
axs[0, 1].set_title('Apps')

axs[1, 0].hist(college["Books"], bins=20, color='orange', alpha=0.7)
axs[1, 0].set_title('Books')

axs[1, 1].hist(college["Personal"], bins=20, color='red', alpha=0.7)
axs[1, 1].set_title('Personal')
```
