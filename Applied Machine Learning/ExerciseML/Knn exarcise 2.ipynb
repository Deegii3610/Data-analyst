```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
np.random.seed(42)
m = 25
Xtrain = np.linspace(0.0,1.0,num=m)
ytrain = 10*Xtrain + np.random.normal(0.0,1.0,m)
Xtrain + np.random.normal(0.0,1.0,m)
Xtrain = Xtrain.reshape(-1, 1)
model = KNeighborsRegressor(n_neighbors=3,weights='uniform').fit(Xtrain, ytrain)
#Test data
Xtest=np.linspace(0.0,1.0,num=1000).reshape(-1, 1)
#Predict on test data
ypred = model.predict(Xtest)
#Plot
import matplotlib.pyplot as plt
plt.rc('font', size=18);
plt.rcParams['figure.constrained_layout.use'] = True
plt.scatter(Xtrain, ytrain, color='red', marker='+')
plt.plot(Xtest, ypred, color='green')
plt.xlabel('input x'); plt.ylabel('output y');
plt.legend(['train','predict'])
plt.show()
```

```python
Xtest1=np.array([0.1, 0.7]).reshape(-1, 1)
#Predict on test data
ypred1 = model.predict(Xtest1)

plt.rc('font', size=18);
plt.rcParams['figure.constrained_layout.use'] = True
plt.scatter(Xtrain, ytrain, color='red', marker='+')
plt.scatter(Xtest1, ypred1, color='green')
plt.xlabel('input x'); plt.ylabel('output y');
plt.legend(['train','predict'])
plt.show()
```

```python
model1 = KNeighborsRegressor(n_neighbors=20,weights='uniform').fit(Xtrain, ytrain)
ypred3 = model1.predict(Xtest)
plt.rc('font', size=18);
plt.rcParams['figure.constrained_layout.use'] = True
plt.scatter(Xtrain, ytrain, color='red', marker='+')
plt.plot(Xtest, ypred3, color='green')
plt.xlabel('input x'); plt.ylabel('output y');
plt.legend(['train','predict'])
plt.show()
```
