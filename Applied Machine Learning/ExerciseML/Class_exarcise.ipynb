# Question 2 

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
y = np.array([140000, 180000, 210000, 250000, 300000])
x = np.array([100000, 200000, 250000, 320000, 355000])

# y= ax+b
b_values = [105, 106, 108]
a_values = [3.7, 3.2, 3.9]

# Plotting the data points
plt.scatter(x, y, color='black', label='Data Points')

# Plotting the linear regression lines
for a, b in zip(a_values, b_values):
    y_pred = a * x + b
    RS = (y - y_pred)**2
    RSS = sum(RS)
    print(f"Residual Sum of Squares {RSS}, when b0 = {b}, b1 = {a}")
    plt.plot(x, y_pred, label=f'y = {a}x + {b}')

x = np.array([100000, 200000, 250000, 320000, 355000]).reshape(-1, 1)
linear_model = LinearRegression().fit(x, y)
ypred_linear = linear_model.predict(x)

plt.plot(x, ypred_linear, color='blue', label='Model Fit Linear regression model')

# Adding labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()


```

```python
import numpy as np

# Given estimates and standard errors
beta0_estimate = 105.73185
beta1_estimate = 3.7889
SE_beta0 = 3.458
SE_beta1 = 0.102376

# Critical value for 95% confidence interval
critical_value = 1.96

# Calculate confidence intervals
CI_beta0_lower = beta0_estimate - critical_value * SE_beta0
CI_beta0_upper = beta0_estimate + critical_value * SE_beta0

CI_beta1_lower = beta1_estimate - critical_value * SE_beta1
CI_beta1_upper = beta1_estimate + critical_value * SE_beta1

# Print the results
print("95% Confidence Interval for beta0: ({:.4f}, {:.4f})".format(CI_beta0_lower, CI_beta0_upper))
print("95% Confidence Interval for beta1: ({:.4f}, {:.4f})".format(CI_beta1_lower, CI_beta1_upper))
```
